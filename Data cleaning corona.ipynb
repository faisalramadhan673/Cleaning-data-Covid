```python
##sekarang print bagian negara
colnames = ['No', 'ObservationDate', 'State', 'Country', 'Update', 'Confirmed', 'Deaths', 'Recovered'] ##inisialisasi nama kolom
covid = pd.read_csv('covid_19_data.csv', names=colnames)
covid = covid.sort_values(by=['Country'])
covid.head()
```

```python
negara = covid.Country.tolist() #simpan kolom Country ke lisst
terjangkit = covid.Confirmed.tolist() #simpan kolom Confirmed ke lisst


output = open("covid2019", "w")
count = 0
value = 0.0


for i,j in zip(negara, terjangkit):
    if count<len(negara)-1 and i == negara[count+1]:
        value += float(j)
        count += 1
        continue
    value += float(j)
    output.write(i + " " + str(value) + "\n")
    
    
    value = 0
    count += 1

output.close()
```
